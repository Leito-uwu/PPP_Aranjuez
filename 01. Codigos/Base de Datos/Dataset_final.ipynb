```python
# ==========================================
# 1. CONFIGURACIÓN DE RUTAS
# ==========================================
# Donde están tus fotos originales sacadas por la cámara
ruta_origen = '/content/drive/MyDrive/PPP/Dataset_Original/' 

# Donde guardaremos las fotos ya recortadas y filtradas para la IA
ruta_destino = '/content/drive/MyDrive/PPP/Dataset_Final/'

# Las categorías que estamos clasificando
categorias = ['Buenas', 'Malas']

print("Iniciando la línea de ensamblaje de datos...")

```

```python
# CELDA 2: Procesamiento Masivo de Botellas

# 1. TUS RUTAS (¡Cámbialas por las tuyas!)
ruta_origen = '/content/drive/MyDrive/PPP/Dataset/Dataset_Original/' 
ruta_destino = '/content/drive/MyDrive/PPP/Dataset/Dataset_Final/'

categorias = ['Botellas_Buenas', 'Botellas_Malas']

print("⚙️ Iniciando la línea de ensamblaje de datos...\n")

# 2. BUCLE PRINCIPAL
for categoria in categorias:
    carpeta_origen = os.path.join(ruta_origen, categoria)
    carpeta_destino = os.path.join(ruta_destino, categoria)
    
    # Escudo Anti-Errores
    if not os.path.exists(carpeta_origen):
        print(f"❌ ERROR: No encuentro la carpeta: '{carpeta_origen}'")
        continue 

    # Crear la carpeta de destino si no existe
    if not os.path.exists(carpeta_destino):
        os.makedirs(carpeta_destino)
        
    archivos = os.listdir(carpeta_origen)
    print(f"Procesando {len(archivos)} imágenes de la categoría: [{categoria}]...")
    
    exitos = 0
    fallos = 0
    
    for nombre_archivo in archivos:
        ruta_img_cruda = os.path.join(carpeta_origen, nombre_archivo)
        
        imagen_original = cv2.imread(ruta_img_cruda)
        if imagen_original is None:
            continue
            
        # El Pipeline de la línea de Aranjuez
        bbox = trabajar_con_contorno_blanco(imagen_original, debug=False)
        
        if bbox is not None:
            recorte = recortar_botella_dinamico(imagen_original, bbox, margen=40)
            
            if recorte is not None:
                imagen_lista_ia = procesar_imagen(recorte)
                
                # Guardamos la imagen final procesada
                ruta_img_final = os.path.join(carpeta_destino, nombre_archivo)
                cv2.imwrite(ruta_img_final, imagen_lista_ia)
                exitos += 1
            else:
                fallos += 1
        else:
            fallos += 1
            
    print(f"   ✅ Éxito: {exitos} | ❌ Fallos: {fallos}\n")

print("🎉 ¡Proceso terminado! Revisa tu carpeta Dataset_Final.")
```
